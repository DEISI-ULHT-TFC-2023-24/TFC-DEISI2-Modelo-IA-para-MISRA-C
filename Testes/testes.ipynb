```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score, precision_score,recall_score, confusion_matrix, accuracy_score, ConfusionMatrixDisplay

# Load predictions
df_gpt4 = pd.read_csv("gpt4_predictions.csv")
df_gpt4o = pd.read_csv("gpt4o_predictions.csv")

# Compute confusion matrices
conf_matrix_gpt4 = confusion_matrix(df_gpt4['annot'], df_gpt4['gpt'])
conf_matrix_gpt4o = confusion_matrix(df_gpt4o['annot'], df_gpt4o['gpt'])
cm_displaygpt4 = ConfusionMatrixDisplay(confusion_matrix = conf_matrix_gpt4, display_labels = [0, 1])
cm_displaygpt4o  = ConfusionMatrixDisplay(confusion_matrix = conf_matrix_gpt4o, display_labels = [0, 1])
print("GPT4")
cm_displaygpt4.plot()
print("GPT4o")
cm_displaygpt4o.plot()


# Calculate precision and recall for GPT-4
TP_gpt4 = conf_matrix_gpt4[1, 1]
FP_gpt4 = conf_matrix_gpt4[0, 1]
FN_gpt4 = conf_matrix_gpt4[1, 0]
TN_gpt4 = conf_matrix_gpt4[0, 0]
precision_gpt4 =  precision_score(df_gpt4['annot'], df_gpt4['gpt'])
recall_gpt4 = recall_score(df_gpt4['annot'], df_gpt4['gpt'])

# Calculate precision and recall for GPT-4O
TP_gpt4o = conf_matrix_gpt4o[1, 1]
FP_gpt4o = conf_matrix_gpt4o[0, 1]
FN_gpt4o = conf_matrix_gpt4o[1, 0]
TN_gpt4o = conf_matrix_gpt4o[0, 0]
precision_gpt4o = precision_score(df_gpt4o['annot'], df_gpt4o['gpt'])
recall_gpt4o = recall_score(df_gpt4o['annot'], df_gpt4o['gpt'])

# Print precision and recall for both models
print(f"GPT-4 precision: {precision_gpt4}, recall: {recall_gpt4}")
print(f"GPT-4O precision: {precision_gpt4o}, recall: {recall_gpt4o}")

# Plot comparison
labels = ['TP', 'FP', 'FN', 'TN', 'Precision', 'Recall']
values_gpt4 = [TP_gpt4, FP_gpt4, FN_gpt4, TN_gpt4, precision_gpt4, recall_gpt4]
values_gpt4o = [TP_gpt4o, FP_gpt4o, FN_gpt4o, TN_gpt4o, precision_gpt4o, recall_gpt4o]

x = range(len(labels))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x, values_gpt4, width, label='GPT-4')
rects2 = ax.bar([p + width for p in x], values_gpt4o, width, label='GPT-4O')

ax.set_xlabel('Metrics')
ax.set_ylabel('Value')
ax.set_title('GPT-4 vs GPT-4O: Metrics Comparison')
ax.set_xticks([p + width / 2 for p in x])
ax.set_xticklabels(labels)
ax.legend()

F1_gpt4= f1_score(df_gpt4['annot'], df_gpt4['gpt'])
F1_gpt4o= f1_score(df_gpt4o['annot'], df_gpt4o['gpt'])


print(f"GPT-4 F1 SCORE: {F1_gpt4}")
print(f"GPT-4O F1 SCORE: {F1_gpt4o}")

accuracy_gpt4= accuracy_score(df_gpt4['annot'], df_gpt4['gpt'])
accuracy_gpt4o= accuracy_score(df_gpt4o['annot'], df_gpt4o['gpt'])

print(f"GPT-4 ACCURACY: {accuracy_gpt4}")
print(f"GPT-4O ACCURACY: {accuracy_gpt4o}")

specificity_gpt4= (TN_gpt4)/(TN_gpt4+FP_gpt4)
specificity_gpt4o= (TN_gpt4o)/(TN_gpt4o+FP_gpt4o)


print(f"GPT-4 SPECIFICITY: {specificity_gpt4}")
print(f"GPT-4O SPECIFICITY: {specificity_gpt4o}")



plt.show()


#id true e o annot

```
